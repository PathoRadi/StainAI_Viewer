```python
import pandas as pd
from PIL import Image, ImageDraw

# Define the file path
xls_path = r'C:\mydjangoapp\stainAI_disp_v001\myapp\static\myapp\demo_Ctrl.xls'
image_path = r'C:\mydjangoapp\stainAI_disp_v001\myapp\static\myapp\demo_Ctrl.jpg'


# Load the Excel file
df = pd.read_excel(xls_path)
img = Image.open(image_path)

bbox_values = df['bbox'].values
# img.show()
overlay = Image.new('RGBA', img.size, (0, 0, 0, 0))

draw = ImageDraw.Draw(overlay)
# Define the color for each value in the 'C50' column
colors = {
    'R': (102, 204, 0, 128),
    'H': (204, 204, 0, 128),
    'B': (220, 112, 0, 128),
    'A': (204, 0, 0, 128),
    'RD': (0, 210, 210, 128),
    'HR': (0, 0, 204, 128)
}

indices = df.groupby('C50').indices


# Convert the 'bbox' column to a string and remove the brackets
df['bbox'] = df['bbox'].str.strip('[]')

# Split the 'bbox' column by spaces and convert each substring to a number
df['bbox'] = df['bbox'].str.split().apply(lambda bbox: [float(num_str) for num_str in bbox])

# Calculate the left, upper, right, and lower coordinates of each bounding box
df[['left', 'upper', 'dx', 'dy']] = pd.DataFrame(df['bbox'].to_list(), index=df.index)
df['right'] = df['left'] + df['dx']
df['lower'] = df['upper'] + df['dy']

# Draw the bounding boxes with the corresponding color
for value, color in colors.items():
    bboxes = df[df['C50'] == value][['left', 'upper', 'right', 'lower']].values
    for bbox in bboxes:
        draw.rectangle(bbox, outline=color, fill=color)
        
from multiprocessing import Pool

# Function to apply alpha_composite on a part of the image
def apply_alpha_composite(args):
    img_part, overlay_part = args
    return Image.alpha_composite(img_part, overlay_part)


# Split the image and the overlay into parts
parts = [(img.crop(box), overlay.crop(box)) for box in bboxes]

# Create a pool of workers
with Pool() as p:
    results = p.map(apply_alpha_composite, parts)

# Combine the results
img_with_boxes = Image.alpha_composite(*results)

# Display the image
img_with_boxes.show()
```
